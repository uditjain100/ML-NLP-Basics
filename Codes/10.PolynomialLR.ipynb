**10. Polynomial Linear Regresssion**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
X = 6*np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
r2_score(y_test, lr.predict(X_test))
```

```python
plt.plot(X_train, lr.predict(X_train), color="red")
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
```

```python
poly = PolynomialFeatures(degree=2)
# poly = PolynomialFeatures(degree=2, include_bias=False) # To remove intercept term i.e, coeff of x^0

# Transformation should be only done on input
X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.fit_transform(X_test)

lr = LinearRegression()
lr.fit(X_train_trans, y_train)
```

```python
r2_score(y_test, lr.predict(X_test_trans))
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new = np.linspace(-3,3,200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)

plt.plot(X_new, y_new, "g.")
plt.plot(X_train, y_train, "b.")
plt.plot(X_test, y_test, "r.")
plt.xlabel("X")
plt.ylabel("y")
```
