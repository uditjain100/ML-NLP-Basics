```python
from sklearn.model_selection import train_test_split

X = df.drop('is_canceled', axis=1).values
y = df['is_canceled'].values * 2 - 1
```

```python
class CustomizedDecisionTree:
    def __init__(self, max_depth=1):
        self.max_depth = max_depth
        self.tree = None

    def fit(self, X, y, depth=0):
        if depth == self.max_depth or len(set(y)) == 1:
            return np.mean(y)

        idx, val = self.best_split(X, y)
        if idx is None:
            return np.mean(y)
        left_mask = X[:, idx] <= val
        right_mask = X[:, idx] > val
        return {
            'index': idx, 'value': val,
            'left': self.fit(X[left_mask], y[left_mask], depth + 1),
            'right': self.fit(X[right_mask], y[right_mask], depth + 1)
        }

    def gini_impurity(self, y):
        classes, counts = np.unique(y, return_counts=True)
        impurity = 1 - sum((count / len(y)) ** 2 for count in counts)
        return impurity

    def best_split(self, X, y):
        best_gini, best_index, best_value = float("inf"), None, None
        for col in range(X.shape[1]):
            values = np.unique(X[:, col])
            for val in values:
                left = y[X[:, col] <= val]
                right = y[X[:, col] > val]
                gini = (len(left) / len(y)) * self.gini_impurity(left) + (len(right) / len(y)) * self.gini_impurity(right)
                if gini < best_gini:
                    best_gini, best_index, best_value = gini, col, val
        return best_index, best_value

    def predict_single(self, x, tree):
        if not isinstance(tree, dict):
            return tree
        idx, val = tree['index'], tree['value']
        branch = 'left' if x[idx] <= val else 'right'
        return self.predict_single(x, tree[branch])

    def predict(self, X):
        return np.array([self.predict_single(x, self.tree) for x in X])
```

```python
class CustomizedRandomForest:
  def __init__(self, n_trees=100):
    self.n_trees = n_trees
    self.trees = []

  def fit(self, X, y):
    for _ in range(self.n_trees):
      indices = np.random.choice(len(X), len(X), replace=True)
      X_sample, y_sample = X[indices], y[indices]
      tree = CustomizedDecisionTree(max_depth=2)
      tree.tree = tree.fit(X_sample, y_sample)
      self.trees.append(tree)

  def predict(self, X):
    predictions = np.array([tree.predict(X) for tree in self.trees])
    return np.round(predictions.mean(axis=0))
```

```python
forest = CustomizedRandomForest(n_trees=5)
forest.fit(X, y)
y_pred = forest.predict(X)
accuracy = np.mean(y_pred == y)
print(f"Accuracy: {accuracy}")
```

```python
import numpy as np

class CustomizedAdaboost:
    def __init__(self, n_trees=100):
        self.n_trees = n_trees
        self.trees = []
        self.alphas = []
        self.weights = None

    def fit(self, X, y):
        self.weights = np.ones(len(X)) / len(X)

        for _ in range(self.n_trees):
            indices = np.random.choice(len(X), len(X), replace=True, p=self.weights)
            X_sample, y_sample = X[indices], y[indices]

            tree = CustomizedDecisionTree(max_depth=1)
            tree.tree = tree.fit(X_sample, y_sample)

            predictions = tree.predict(X)
            incorrect = (predictions != y).astype(int)

            error = np.dot(self.weights, incorrect) / np.sum(self.weights)
            alpha = 0.5 * np.log((1 - error) / (error + 1e-10))

            self.weights *= np.exp(alpha * incorrect)
            self.weights /= np.sum(self.weights)

            self.alphas.append(alpha)
            self.trees.append(tree)

    def adaboost_predict(self, X):
        tree_preds = np.array([alpha * tree.predict(X) for tree, alpha in zip(self.trees, self.alphas)])
        return np.round(np.sign(tree_preds.sum(axis=0)))
```

```python
class CustomizedGradientBoosting:
    def __init__(self, n_estimators=10, learning_rate=0.1, max_depth=1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.trees = []
        self.initial_prediction = None

    def fit(self, X, y):
        self.initial_prediction = np.mean(y)
        current_pred = np.full(y.shape, self.initial_prediction)

        for _ in range(self.n_estimators):
            residuals = y - current_pred

            tree = CustomizedDecisionTree(max_depth=self.max_depth)
            tree.tree = tree.fit(X, residuals)

            tree_pred = tree.predict(X)
            current_pred += self.learning_rate * tree_pred

            self.trees.append(tree)

    def predict(self, X):
        pred = np.full(X.shape[0], self.initial_prediction)

        for tree in self.trees:
            pred += self.learning_rate * tree.predict(X)
        return np.sign(pred)
```
