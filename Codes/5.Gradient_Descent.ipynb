**5. Gradient Descent**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression

X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
plt.scatter(X,y)

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# OLS

reg = LinearRegression()
reg.fit(X, y)

y_pred = reg.predict(X)

print(reg.intercept_)
print(reg.coef_)

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
```

```python
# Gradient Descent for calculating intercept (c) value with initial value = 0 and with m = 78.5 (constant)

m = 78.5
c = 0

y_pred = ((m * X) + c).reshape(4)

plt.scatter(X, y)

# OLS
plt.plot(X, reg.predict(X), color='red', label="OLS")

# Gradient Descent Initial
plt.plot(X, y_pred, color='#00a65a', label="GD0")
```

```python
# value of slope, learning rate and stepsize

slope = -2 * (np.sum(y- m*X.ravel() - c))
print(slope)

lr = 0.1

ss = lr * slope
print(ss)

c = c - ss

y_pred = ((m * X) + c).reshape(4)

plt.scatter(X, y)

# OLS
plt.plot(X, reg.predict(X), color='red', label="OLS")

# Gradient Descent 1st iteration
plt.plot(X, y_pred, color='#00a65a', label="GD1")
```

```python
slope = -2 * (np.sum(y- m*X.ravel() - c))
ss = lr * slope
c = c - ss

y_pred = ((m * X) + c).reshape(4)

plt.scatter(X, y)

# OLS
plt.plot(X, reg.predict(X), color='red', label="OLS")

# Gradient Descent 2nd iteration
plt.plot(X, y_pred, color='#00a65a', label="GD2")
```

```python
epocs = 10

plt.scatter(X, y)

# OLS
plt.plot(X, reg.predict(X), color='red', label="OLS")

for i in range(epocs):
  slope = -2 * (np.sum(y- m*X.ravel() - c))
  ss = lr * slope
  c = c - ss

  y_pred = ((m * X) + c).reshape(4)

  # Gradient Descent 2nd iteration
  plt.plot(X, y_pred, color='#00a65a', label="GD{}".format(i))
```
