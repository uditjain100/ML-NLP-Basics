```python
# !pip install mne
import mne
import pandas as pd

raw = mne.io.read_raw_gdf("/content/drive/MyDrive/A01E.gdf", preload=True)
raw
type(raw)
```

```python
raw.info
raw.ch_names
type(raw.get_data())
```

```python
data = raw.get_data()
times = raw.times

df_ch = pd.DataFrame(data.T, columns=raw.ch_names)
df_ch

df_ch['time'] = times
df_ch
```

```python
events, event_id = mne.events_from_annotations(raw)
df_ev = pd.DataFrame(events, columns=['time', 'duration', 'description'])
df_ev
```

```python
missing_data_ch = df_ch.isna().sum()
missing_data_ch

missing_data_ev = df_ev.isna().sum()
missing_data_ev
```

```python
from scipy.stats import skew
ch_skewness = df_ch.apply(lambda x: skew(x))
ch_skewness

```

```python
from sklearn import preprocessing
import numpy as np
df_ch_columns = df_ch.columns

# scaler = preprocessing.StandardScaler()
# df_ch = scaler.fit_transform(df_ch)
# df_ch = pd.DataFrame(df_ch, columns=df_ch_columns)
# df_ch

from sklearn.preprocessing import PowerTransformer

scaler = PowerTransformer(method='yeo-johnson', standardize=False)
df_ch = scaler.fit_transform(df_ch)
df_ch = pd.DataFrame(df_ch, columns=df_ch_columns)
df_ch
```

```python
ch_skewness = df_ch.apply(lambda x: skew(x))
ch_skewness
```

```python
from sklearn.model_selection import train_test_split

labels = df_ch['EEG-Fz']
df_ch = df_ch.drop('EEG-Fz', axis=1)

X_train, X_test, y_train, y_test = train_test_split(df_ch, labels, test_size=0.3, random_state=42)
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
import matplotlib.pyplot as plt

plt.plot(y_test, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.title('Actual vs Predicted')
plt.show()
```
