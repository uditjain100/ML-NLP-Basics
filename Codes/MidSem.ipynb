```python
class MLR:

  def __init__(self):
    self.coef = None
    self.intercept = None

  def fit(self, X, y):
    x = np.insert(X, 0, 1, axis=1)

    betas  = np.linalg.inv(np.dot(x.T, x)).dot(x.T).dot(y)
    self.intercept = betas[0]
    self.coef = betas[1:]

  def predict(self, X):
    return np.dot(X, self.coef) + self.intercept
```

```python
from sklearn.metrics import r2_score

mlr = MLR()
mlr.fit(X_train, y_train)
print(r2_score(y_test, mlr.predict(X_test)))
```

```python
class BGD:
  def __init__(self):
    self.intercept = 0
    self.coeff = None
    self.learning_rate = 0.1
    self.epocs = 100

  def fit(self, X, y):
    self.coeff = np.ones(X.shape[1])
    for i in range(self.epocs):

      print(X.shape, self.coeff.shape)
      y_pred = self.intercept + np.dot(X, self.coeff)

      slope_int = -2 * np.mean(y - y_pred)
      self.intercept -= self.learning_rate * slope_int

      slope_coeff = -2 * np.dot(y-y_pred, X) / X.shape[0]
      self.coeff -= self.learning_rate * slope_coeff

  def predict(self, X):
    return np.dot(X, self.coeff) + self.intercept
```

```python
class SGD:
  def __init__(self):
    self.intercept = 0
    self.coeff = None
    self.learning_rate = 0.1
    self.epocs = 100

  def fit(self, X, y):
    self.coeff = np.ones(X.shape[1])
    for i in range(self.epocs):
      for j in range(X.shape[0]):
        idx = np.random.randint(0, X.shape[0])

        y_pred = self.intercept + np.dot(X[idx], self.coeff)

        slope_coeff = -2 * y[idx] - y_pred
        self.coeff -= self.learning_rate * slope_coeff

        slope_int = -2 * np.dot(y[idx]-y_pred, X[idx])
        self.intercept -= self.learning_rate * slope_int

  def predict(self, X):
    return np.dot(X, self.coeff) + self.intercept
```

```python
class MBGD:
  def __init__(self):
    self.intercept = 0
    self.coeff = None
    self.learning_rate = 0.1
    self.epocs = 100
    self.batch_size = 5

  def fit(self, X, y):
    self.coeff = np.ones(X.shape[1])
    for i in range(self.epocs):
      for j in range(X.shape[0] / self.batch_size):
        idx = np.sample(range(X.shape(0)), self.batch_size)

        y_pred = self.intercept + np.dot(X[idx], self.coeff)

        slope_coeff = -2 * np.mean(y[idx] - y_pred)
        self.coeff -= self.learning_rate * slope_coeff

        slope_int = -2 * np.dot(y[idx]-y_pred, X[idx])
        self.intercept -= self.learning_rate * slope_int

  def predict(self, X):
    return np.dot(X, self.coeff) + self.intercept
```

```python
class DT:

  def __init(self):
    self.max_depth = 1
    self.tree = None

  def fit(self, X, y, depth=1):
    if depth == self.max_depth or len(set(y)) == 1:
      return np.mean(y)

    idx, val = self._best_split(X, y)
    if idx == None:
      return np.mean(y)

    left = X[:, idx] <= val
    right = X[:, idx] > val

    return { 'index': idx, 'value': val, 'left': self.fit(X[left], y[left], depth + 1), 'right': self.fit(X[right], y[right], depth + 1) }

  def _best_split(self, X, y):
    best_gini, best_value, best_index = float('inf'), None, None

    for col in X.shape[1]:
      values = np.unique(X[:, col])
      for val in values:
        left = y[X[:, col] <= val]
        right = y[X[:, col] > val]
        gini = (len(left)/len(y)) * self._gini_impurity(left) + (len(right)/len(y)) * self._gini_impurity(right)
        if gini < best_gini:
          best_gini = gini
          best_value = val
          best_index = col
    return best_index, best_value

  def _gini_impurity(self, y):
    classes, counts = np.unique(y, return_counts=True)
    impurity = 1 - sum((count / len(y)) ** 2 for count in counts)
    return impurity

  def predict(self, X):
    return np.array([self._predict_single(x, self.tree) for x in X])

  def _predict_single(self, x, tree):
    if not isinstance(tree, dict):
      return tree

    idx, val = tree['index'], tree['value']
    branch = 'left' if x[idx] <= val else 'right'
    return self.predict_single(x, tree[branch])
```

```python
class RF:

  def __init__(self):
    self.n_trees = 5
    self.trees = []

  def fit(self, X, y):
    index = np.random.choice(len(X), len(X), inplace=True)

    xSample = X[index]
    ySample = y[index]

    dt = DT()
    dt.tree = dt.fit(xSample, ySample)
    self.trees.append(dt.tree)

  def predict(self, X):
    y_pred = np.array([tree.predict(X) for tree in self.trees])
    return np.round(y_pred.mean(axis=0))
```

```python
class AB:

  def __init__(self):
    self.n_trees = 5
    self.trees = []
    self.weights = []
    self.alphas = []

  def fit(self, X, y):
    index = np.random.choice(len(X), len(X), inplace=True)

    xSample = X[index]
    ySample = y[index]

    dt = DT()
    dt.tree = dt.fit(xSample, ySample)

    pred = dt.predict(xSample)
    incorrect = (pred != y).astype(int)

    error = np.dot(self.weights, incorrect) / np.sum(self.weights)
    alpha = (0.5) * np.log((1 - error) / (error + 1e-10))

    self.weights *= np.exp(error)
    self.weights /= np.sum(self.weights)

    self.alphas.append(alpha)
    self.trees.append(dt.tree)

  def predict(self, X):
    y_pred = np.array([alpha * tree.predict(X) for tree, alpha in zip(self.trees, self.alphas)])
    return np.round(np.sign(y_pred.sum(axis=0)))
```

```python
class GB:
  def __init__(self):
    self.n_trees = 5
    self.trees = []
    self.learning_rate = 0.1
    self.max_depth = 5
    self.initial_pred = None

  def fit(self, X, y):
    self.initial_prediction = np.mean(y)
    curr_pred = np.full(y.shape, self.initial_prediction)

    for i in range(self.n_trees):
      res = y - curr_pred

      dt = DT()
      dt.tree = dt.fit(X, res)
      curr_pred += self.learning_rate * dt.pred(X)

      self.trees.append(dt)

  def predict(self, X):
    pred = np.full(X.shape[0], self.initial_prediction)

    for tree in self.trees:
      pred += self.learning_rate * tree.predict(X)

    return np.sign(pred)
```

```python
class LGR:

  def __init__(self):
    self.W = None
    self.learning_rate  = 0.1
    self.epoch = 100

  def softmax(self, X):
    num = np.exp(X)
    den = np.sum(num, axis=1, keepdims=True)
    return num / den

  def fit(self, X, y):
    self.W = np.random.randn(X.shape[1], y.shape[1])

    for i in range(self.epoch):
      eq = np.dot(X, self.W)
      y_pred = self.softmax(eq)
      self.W = self.W - self.learning_rate * ((1/X.shape[0]) * np.dot(X.T, y_pred - y))

  def predict(self, X):
    eq = np.dot(X, self.W)
    y_pred = self.softmax(eq)
    return np.argmax(y_pred, axis=1)
```
