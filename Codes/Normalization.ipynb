```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
import statistics

data = pd.read_csv("/content/titanic.csv")
data = pd.DataFrame(data['PassengerId'], columns=['PassengerId'])
data.head()
```

```python
# 1. Z Score

# Manually
meanPassengerId = data["PassengerId"].mean()
stDevPassengerId = statistics.stdev(data["PassengerId"])

def getNewPID(ele):
  return float(ele - meanPassengerId) / float(stDevPassengerId)

passengerIdNew = data["PassengerId"].apply(lambda ele : getNewPID(ele))

# # Using Library
scaler = preprocessing.StandardScaler()
passengerIdNew = scaler.fit_transform(data)

for i in range(0, len(data["PassengerId"])):
  print(passengerIdNew[i], data["PassengerId"][i], sep=" - ")
```

```python
# 2. Min-Max Scaler

# Manually
minPassengerId = data["PassengerId"].min()
maxPassengerId = data["PassengerId"].max()

def getNewPassengerId(ele):
  return float(ele - minPassengerId) / float(maxPassengerId - minPassengerId)

PassengerIdNew = data["PassengerId"].apply(lambda ele : getNewPassengerId(ele))

# Using Library
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
PassengerIdNew = scaler.fit_transform(data)

for i in range(0, len(data["PassengerId"])):
  print(PassengerIdNew[i], data["PassengerId"][i], sep=" - ")
```

```python
# 3. Max Absolute Scaling

# Manually
maxPassengerId = data["PassengerId"].max()

def getNewPassengerId(ele):
  return float(ele) / float(maxPassengerId)

PassengerIdNew = data["PassengerId"].apply(lambda ele : getNewPassengerId(ele))

# Using Library
scaler = preprocessing.MaxAbsScaler()
PassengerIdNew = scaler.fit_transform(data)

for i in range(0, len(data["PassengerId"])):
  print(PassengerIdNew[i], data["PassengerId"][i], sep=" - ")
```

```python
# 4. Robust Scaling

# Manually
medianPassengerId = data["PassengerId"].median()

q1 = np.percentile(data["PassengerId"], 25)
q3 = np.percentile(data["PassengerId"], 75)
iqr = q3 - q1

def getNewPID(ele):
    return float(ele - medianPassengerId) / float(iqr)

passengerIdNew = data["PassengerId"].apply(lambda ele : getNewPID(ele))

# Using Library
scaler = preprocessing.RobustScaler()
passengerIdNew = scaler.fit_transform(data)

for i in range(len(data["PassengerId"])):
  print(passengerIdNew[i], data["PassengerId"][i], sep=" - ")
```
