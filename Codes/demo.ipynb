```python
import pandas as pd
```

```python
import io
data = pd.read_csv(io.BytesIO(uploaded["titanic (2).csv"]))
```

```python
data.head(50)
```

```python
missing = data.isnull().sum()
missing
```

```python
data["Age"].fillna(data["Age"].median(), inplace=True)
data.head(50)
```

```python
data['Fare'].fillna(data['Fare'].median(), inplace=True)
data
```

```python
data["CabinExist"] = data["Cabin"].apply(lambda ele : 0 if pd.isna(ele) else 1)
data.head(50)
```

```python
data["Sex"] = data["Sex"].map({"male" : 0, "female" : 1})
data.head(10)
```

```python
data["Embarked"] = data["Embarked"].map({ "C":0, "Q":1, "S":2 })
data.head(10)
```

```python
data["FamilyStatus"] = data.apply(lambda row : 1 if row["SibSp"] > 0 or row["Parch"] > 0 else 0, axis = 1)
data.head(10)
```

```python
fare_median = data["Fare"].median()
data["HighFare"] = data["Fare"].apply(lambda f: 1 if f > fare_median else 0)
data.head(10)
```

```python
bins = [0, 10, 18, 40, 70, 100]
labels = [0, 1, 2, 3, 4]
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels)
data.head(10)
```

```python
survival_data = data[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'CabinExist']]
class_data = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'CabinExist']]
gender_data = data[['Sex', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'CabinExist']]
embarked_data = data[['Embarked', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'CabinExist']]
cabin_data = data[['CabinExist', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
family_status_data = data[['FamilyStatus', 'Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'CabinExist']]
high_fare_data = data[['HighFare', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked', 'CabinExist']]
age_group_data = data[['AgeGroup', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Fare', 'Embarked', 'CabinExist']]

updatedData = {
    'Survival': survival_data.head(),
    'Class': class_data.head(),
    'Gender': gender_data.head(),
    'Embarkation Port': embarked_data.head(),
    'Cabin': cabin_data.head(),
    'Family Status': family_status_data.head(),
    'High-Fare': high_fare_data.head(),
    'Age Group': age_group_data.head()
}

updatedData
```
