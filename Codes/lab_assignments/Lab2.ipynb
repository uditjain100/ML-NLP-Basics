```python
# Importing Dataset
import pandas as pd
data = pd.read_excel("/content/drive/MyDrive/Lab/ML/Lab_2/titanic.xlsx")
data.head(10)
```

```python
missing_data = data.isna().sum()
missing_data
```

```python
# Age Null Values Handling
mean_age = data["Age"].mean()
mean_age = int(mean_age)

data["Age"] = data["Age"].apply(lambda ele : mean_age if pd.isna(ele) else ele)
data.head(50)
```

```python
# Fare Null Values Handling
mean_fare = data["Fare"].mean()
mean_fare = int(mean_fare)

data["Fare"] = data["Fare"].apply(lambda ele : mean_fare if pd.isna(ele) else ele)
data.head(20)
```

```python
# HasCabin Label Encoding
data["HasCabin"] = data["Cabin"].apply(lambda ele : 0 if pd.isna(ele) else 1)
data.head(20)
```

```python
# Sex Label Encoding
# male -> 0
# female -> 1

# Manual
# data["Sex"] = data["Sex"].map(lambda ele : 0 if ele == 'male' else 1)
# data.head(10)

from sklearn import preprocessing

# Using Label Encoder
# labelEncoder = preprocessing.LabelEncoder()
# data["Sex"] = labelEncoder.fit_transform(data["Sex"])
# data.head(15)

# Using One hot Encoder
oneHotEncoder = preprocessing.OneHotEncoder(sparse_output=False)
encodedData = oneHotEncoder.fit_transform(data[["Sex"]])
encodedData

encodedData = pd.DataFrame(encodedData, columns=oneHotEncoder.get_feature_names_out(["Sex"]))
encodedData

data = pd.concat([data, encodedData], axis=1)
data = data.drop(['Sex'], axis=1)
data
```

```python
# Embarked Label Encoding

# Manual
# C -> 0, Q -> 1, s -> 2
# data["Embarked"] = data["Embarked"].map({ 'C' : 0, 'Q' : 1, 'S' : 2 })
# data.head(10)

# Using Label Encoder
labelEncoder = preprocessing.LabelEncoder()
data["Embarked"] = labelEncoder.fit_transform(data["Embarked"])
data.head(15)
```

```python
# Family Status Label Encoding
data["FamilyStatus"] = data.apply(lambda row : 1 if row["SibSp"] > 0 or row["Parch"] > 0 else 0, axis = 1)
data.head(10)
```

```python
# HighFare Label Encoding
data["HighFare"] = data.apply(lambda row : 1 if row["Fare"] > mean_fare else 0, axis = 1)
data.head(10)
```

```python
# Age Group Label Encoding - Bining Method
max_age = data["Age"].max() # 76.0
min_age = data["Age"].min() # 0.17

bins = [0, 10, 18, 40, 70, 100]
labels = [0, 1, 2, 3, 4]
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels)


data.head(10)
```

```python
# Name Label Encoding
uniqueNames = data["Name"].unique()
nameNumberMap = {}
for i, name in enumerate(uniqueNames):
  nameNumberMap[name] = i

data["Name"] = data["Name"].map(nameNumberMap)
data.head(10)
```

```python
# Ticket Label Encoding
tickets = data["Ticket"].unique()
ticketNumberMap = {}
for i, ticket in enumerate(tickets):
  ticketNumberMap[ticket] = i

data["Ticket"] = data["Ticket"].map(ticketNumberMap)
data.head(10)
```

```python
# Cabin Label Encoding
data["Cabin"] = data["Cabin"].map(lambda ele : "Unknown" if pd.isna(ele) else ele)

data.head(15)

uniqueCabins = data["Cabin"].unique()
cabinNumberMap = {}
for i, cabin in enumerate(uniqueCabins):
  cabinNumberMap[cabin] = i

data["Cabin"] = data["Cabin"].map(cabinNumberMap)
data
```
